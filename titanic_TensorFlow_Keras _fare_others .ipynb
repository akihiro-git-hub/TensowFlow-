```python
import tensorflow as tf
from tensorflow.keras import models, layers 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_train = pd.read_csv('x_train_fare_others.csv')
y_train = pd.read_csv('y_train_fare_others.csv')
print(type(x_train))
print(type(y_train))
print(x_train.shape)
print(y_train.shape)
```

```python
input_num = 5
midle_units = 3
num_class = 2

y_train = tf.keras.utils.to_categorical(y_train,num_class)
print(y_train.shape)

model = tf.keras.Sequential()
model.add(layers.Dense(activation = 'relu',input_dim = input_num, units = midle_units, name = 'layer1')),
model.add(layers.Dropout(0.2))
model.add(layers.Dense(activation='relu',units = midle_units ,name = 'layer2'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(activation='relu',units = midle_units ,name = 'layer3'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(activation = 'sigmoid',units = num_class ,name = 'layer4'))
```

```python
model.compile(loss =  'categorical_crossentropy', optimizer='RMSprop',metrics =['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train, epochs=1000, verbose=1, validation_split=0.1)
```
