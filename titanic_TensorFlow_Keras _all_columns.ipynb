```python
import tensorflow as tf
from tensorflow.keras import models, layers 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x_train = pd.read_csv('x_train_all_columns.csv')
y_train = pd.read_csv('y_train_all_columns.csv')
print(type(x_train))
print(type(y_train))
print(x_train.shape)
print(y_train.shape)
```

```python
input_num = 6
midle_units = 10
num_class = 2

y_train = tf.keras.utils.to_categorical(y_train,num_class)
print(y_train.shape)

model = tf.keras.Sequential()
model.add(layers.Dense(activation = 'relu',input_dim = input_num, units = midle_units, name = 'layer1')),
model.add(layers.Dropout(0.15))
model.add(layers.Dense(activation='relu',units = midle_units ,name = 'layer2'))
model.add(layers.Dropout(0.15))
model.add(layers.Dense(activation='relu',units = midle_units ,name = 'layer3'))
model.add(layers.Dropout(0.15))
model.add(layers.Dense(activation = 'sigmoid',units = num_class ,name = 'layer4'))
```

```python
model.compile(loss =  'categorical_crossentropy', optimizer='RMSprop',metrics =['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x_train,y_train, epochs=10000, verbose=1, validation_split=0.1)
```

```python
#＋2000して3千回まで回数を増やしてみる。
history = model.fit(x_train,y_train, epochs=2000, verbose=1, validation_split=0.1)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(acc))


plt.figure(figsize=(10,10))
plt.plot(epochs, acc,  label = 'train acc')
plt.plot(epochs, val_acc, label= 'valid acc')
plt.plot(epochs, loss, label = 'train loss')
plt.plot(epochs, val_loss,label= 'valid loss')
plt.legend()

plt.title('acc and loss from all columns ')
plt.grid()
plt.ylim([0.3,1.0])
plt.savefig('all_columns_acc_and_loss_units10_Dropout15_10000')
plt.show()
```

```python
open('all_columns_unit10_Droopout15_10000.json','w').write(model.to_json())
```
